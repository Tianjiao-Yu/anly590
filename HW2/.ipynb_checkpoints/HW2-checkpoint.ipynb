## 1.3 forward-pass function in python:

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

np.warnings.filterwarnings('ignore', category=np.VisibleDeprecationWarning)
```

```python
def relu(z):
  return np.maximum(0, z)
def sigmoid(z):
  return 1/(1 + np.exp(-z))
```

```python
def ff_nn_relu(x,W,V,b,c):
  h = relu(np.dot(x, W)+b)
  y = sigmoid(np.dot(h, V)+c)
  return y
```

## 2.2 Create a visualization


```python
delta = 0.05
x = np.arange(-10.0, 10.0, delta)
y = np.arange(-10.0, 10.0, delta)
X, Y = np.meshgrid(x, y)
Z = (3-X**3)+50*(2*Y**2-X)**2
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
```

 ## 2.3 Write my grad:

```python
def grad_f(vector):
    x, y = vector
    df_dx = -3*x**2 + 200*y**2 +100*x
    df_dy = 800*y**3 - 400*x*y
    return np.array([df_dx, df_dy])
```

```python
def grad_descent(starting_point, iterations, learning_rate):
  if starting_point is None:
    point = np.random.uniform(-10,10, size=2)
  else: 
    point = starting_point
  return_list = [point]
  for i in range(iterations):
    grad = grad_f(point)
    point = point - learning_rate*grad
    return_list.append(point)
  return np.array(return_list)
```

```python
np.random.seed(5)
traj = grad_descent(None, 10000, 0.00001)

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
x= traj[:,0]
y= traj[:,1]
plt.plot(x,y,'-o')
```

```python
np.random.seed(5)
traj = grad_descent(None, 10000, 0.00002)

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
x= traj[:,0]
y= traj[:,1]
plt.plot(x,y,'-o')
```

```python
np.random.seed(5)
traj = grad_descent(None, 10000, 0.0000001)

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
x= traj[:,0]
y= traj[:,1]
plt.plot(x,y,'-o')
```

## 2.4 Grad with Momentum

```python
def grad_descent_with_momentum(starting_point=None, iterations=10, alpha=.9, epsilon=10):
  if starting_point is None:
    point = np.random.uniform(-10,10, size=2)
  else: 
    point = starting_point
  return_list = [point]
  v = np.zeros(point.size)
  for i in range(iterations):
    grad = grad_f(point)
    v = alpha*v + epsilon*grad
    point = point - v
    return_list.append(point)
  return np.array(return_list)
```

```python
np.random.seed(10)
traj = grad_descent_with_momentum(iterations=10000, epsilon=0.00001, alpha=.8)

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
x= traj[:,0]
y= traj[:,1]
plt.plot(x,y,'-o')
```

```python
np.random.seed(10)
traj = grad_descent_with_momentum(iterations=10000, epsilon=0.000005, alpha=.8)

fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
x= traj[:,0]
y= traj[:,1]
plt.plot(x,y,'-o')
```

## 3.2. Write the function 

```python

def relu(z):
    return np.maximum(0, z)
def sigmoid(z):
    return 1 / (1 + np.exp(-np.array(z)))
def ff_nn_ReLu(X, W, b, V, c):
    res = relu(np.array(X)@np.array(W)+b)
    return sigmoid(np.dot(res,np.array(V))+c)
def grad_f(x, y, y_hat, w, b, v, c):
    loss_y_hat = (y/y_hat - (1-y)/(1-y_hat))
    h = relu(np.array(x)@np.array(w)+b)
    
    dw = x.T.dot(np.expand_dims((loss_y_hat * y_hat*(1-y_hat)),1).dot(np.expand_dims(v,1).T))
    db = (loss_y_hat* y_hat*(1-y_hat)).T.dot(np.where(h<=0,0,1))*v
    
    dv = (loss_y_hat* y_hat*(1-y_hat)).T.dot(h)
    dc = loss_y_hat.dot(y_hat*(1-y_hat))
    
    return np.array([dw,db,dv,dc])/len(y)


x = np.array([[1,-1], [0,-1], [0, 1]])
w = np.array([[1, -1, 0, 1], [0, 0, 0.5, 1]])
b = np.array([0, 0, 1, 0])
v = np.array([1, 0, -1, 1])
c = 1
y_pred = ff_nn_ReLu(x,w,b,v,c)
```

## 3.4. Fit your network

```python
def loss(y, y_hat):
    # binary cross entropy
    return  np.sum(- (y*np.log(y_hat) + (1-y)*np.log(1 - y_hat)))

def grad_descent(x, y_true, w, b, v, c, iterations=10, learning_rate=0.1):  
    y_pred = ff_nn_ReLu(x,w,b,v,c)
    losses = [loss(y_true, y_pred)]
    for i in range(iterations):
        dw,db,dv,dc = grad_f(x, y_true, y_pred, w, b, v, c)
        w = w+learning_rate*dw
        b = b+learning_rate*db
        v = v+learning_rate*dv
        c = c+learning_rate*dc
        y_pred = ff_nn_ReLu(x,w,b,v,c)
        l = loss(y_true, y_pred)
        losses.append(l)
    return np.array([w,b,v,c]),losses
```

```python
array, losses = grad_descent(X, y, w, b,v,c)
```

```python
plt.plot(losses)
plt.ylabel("loss")
plt.xlabel("iterations")
plt.show()
```

## 5. repeat with momentum

```python
def grad_descent_w_momentum(x, y_true, w, b, v, c, iterations=10, alpha=.1, epsilon=0.5):  
    y_pred = ff_nn_ReLu(x,w,b,v,c)
    losses = [loss(y_true, y_pred)]
    vw = np.zeros(w.shape)
    vb = np.zeros(b.shape)
    vv = np.zeros(v.shape)
    vc = 0
    for _ in range(iterations):
        dw,db,dv,dc = grad_f(x, y_true, y_pred, w, b, v, c)
        vw = alpha*vw - epsilon*dw
        vb = alpha*vb - epsilon*db
        vv = alpha*vv - epsilon*dv
        vc = alpha*vc - epsilon*dc
        w = w-vw
        b = b-vb
        v = v-vv
        c = c-vc
        y_pred = ff_nn_ReLu(x,w,b,v,c)
        l = loss(y_true, y_pred)
        losses.append(l)
    return np.array([w,b,v,c]),losses
```

```python
array_m, losses_m =  grad_descent_w_momentum(X, y, w,b,v,c)
```

```python
plt.plot(losses_m)
plt.ylabel("loss")
plt.xlabel("iterations")
plt.show()
```
